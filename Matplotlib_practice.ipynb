```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)

#plt.plot() is to create a figure with current axes. 
# 
```

```python
# When switch from MATLAB to object-oriented interface method
# Create the plot with ax.plt()

fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

# To created a figure, use: fig, ax = plt.subplots()
# To define the axes, use: ax.plot().
# Inside the parentheses of ax.plot(), add the axes separated by a comma.

# plt.subplots() returns a tuple that contains a figure and axes objects
# fig, ax = plt.subplots() unpacks the tuple in to a variable fig and ax
# fit, reference the figure
# ax, reference the axes of the figure and allows us to annotate the figure
# ax.plot(x_axis, y_axis), plots the data from x and y axes
```

```python
# Create the plot with ax.plt()
# result empty chart
fig = plt.figure()
ax = fig.add_subplot()

# add this, will get the same result as using MATLAB method
ax.plot(x_axis, y_axis)
```

```python
# Object-oriented interface method
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#The length of the tuple should be equal to the number of wedges in the pie chart.
# 0.2 : To pop out the seventh month, July
# autopct =:To add the percentage of each wedge on the pie chart, 
# plt.subplots(figsize=(8, 8)): increase size of bar
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
